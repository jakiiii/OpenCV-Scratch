```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython import get_ipython
ipy = get_ipython()
if ipy is not None:
    ipy.run_line_magic('matplotlib', 'inline')
```

```python
img = cv2.imread('D:\Dev\OpenCV\images\cat_face.jpg')
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edge)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WHICHEVER IS GRATER
lower = int(max(0, 0.7 * med_val))
# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OF THE MAX 255, WHICHEVER IS SMALLER
upper = int(max(255, 1.3 * med_val))
```

```python
edge = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

```python
blurred_img = cv2.blur(img, ksize=(5, 5))
```

```python
edge = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

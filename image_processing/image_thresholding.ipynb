```python
img = cv2.imread('D:\Dev\OpenCV\images\crossword.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_img(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_img(img)
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_img(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_img(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_img(blended)
```
